## Prepare the data

```python
y = train['SalePrice']
train.drop("SalePrice", axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder

low_to_high = (['Utilities', 'LandSlope', 'ExterQual','ExterCond', 'BsmtQual', 
                'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 
                'HeatingQC' , 'KitchenQual', 'Functional', 'FireplaceQu' , 'GarageType' , 
               'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence',
                'Street', 'Alley', 'CentralAir'] )

for c in low_to_high :
    lbl = LabelEncoder()
    lbl.fit(list(train[c].values))
    train[c] = lbl.transform(list(train[c].values))
```

Mange av objektene i low_to_high har verdier som "no garage" på "garage quality", som ikke er like lett å kvantifisere på en skala som forteller om hvor god kvaliteten er. Det vi kunne gjort her er å sette verdier som "no garage" til medianverdien på attributten. Vi velger derimot å definere verdier som "no garage" som dårligere enn "bad quality", på alle liknende attributter.

```python
train = pd.get_dummies(train)
```

# 5. Explore many different models and shortlist the best ones

```python
#making train and validation

X_train, X_test, y_train, y_test = train_test_split(train, y, 
                                                    test_size = 0.3, random_state = 42)
print("X_train: ", str(X_train.shape))
print("X_test: ", str(X_test.shape))
print("y_train: ", str(y_train.shape))
print("y_test: ", str(y_test.shape))
```

```python
# Define error measure for official scoring : RMSE
scorer = make_scorer(mean_squared_error, greater_is_better = False)

def rmse_cv(model):
    rmse = np.sqrt(-cross_val_score(model, X_train, 
                                    y_train, scoring = scorer, cv = 5))
    return(rmse)
    
#def rmse_cv_test(model):
#    rmse = np.sqrt(-cross_val_score(model, X_test, 
 #                                   y_test, scoring = scorer, cv = 5))
#    return(rmse)
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, make_scorer

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

print("RMSE: ", rmse_cv(lin_reg).mean())
#print("RMSE on test set: ", rmse_cv_test(lin_reg).mean())
```

```python

# Look at predictions on training and validation set
print("RMSE:", rmse_cv(lin_reg).mean())
#print("RMSE on Test set :", rmse_cv_test(lin_reg).mean())
y_train_pred = lin_reg.predict(X_train)
y_test_pred = lin_reg.predict(X_test)

# Plot residuals
plt.scatter(y_train_pred, y_train_pred - y_train, c = "blue", marker = "s", label = "Training data")
plt.scatter(y_test_pred, y_test_pred - y_test, c = "lightgreen", marker = "s", label = "Validation data")
plt.title("Linear regression")
plt.xlabel("Predicted values")
plt.ylabel("Residuals")
plt.legend(loc = "upper left")
plt.hlines(y = 0, xmin = 10.5, xmax = 13.5, color = "red")
plt.show()

# Plot predictions
plt.scatter(y_train_pred, y_train, c = "blue", marker = "s", label = "Training data")
plt.scatter(y_test_pred, y_test, c = "lightgreen", marker = "s", label = "Validation data")
plt.title("Linear regression")
plt.xlabel("Predicted values")
plt.ylabel("Real values")
plt.legend(loc = "upper left")
plt.plot([10.5, 13.5], [10.5, 13.5], c = "red")
plt.show()
```

#### Lasso

Lasso stands for "Least Absolute Shrinkage and Selection Operator". 

```python
from sklearn.linear_model import Ridge, RidgeCV, ElasticNet, LassoCV, LassoLarsCV
from sklearn.model_selection import cross_val_score

lasso_mod = LassoCV(alphas = [1, 0.6, 0.3, 0.1, 0.001, 0.001, 0.0005], max_iter = 50000, cv = 10)
lasso_mod.fit(X_train, y_train)
alpha = lasso_mod.alpha_
print("best alpha: ", alpha)
rmse_cv(lasso_mod).mean()
```

```python
coef = pd.Series(lasso_mod.coef_, index = X_train.columns)
print("Lasso picked", str(sum(coef != 0 )), " variables and eliminated ", str(sum(coef == 0)), " variables")
```
